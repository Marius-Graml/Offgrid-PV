## Raw voltage and current data

```python
# Get voltage and current data

def generate_voltage_and_current_csv():
    myclient = pymongo.MongoClient(<key>) #insert key for database here
    mydb = myclient["offgridpv"]
    mycol = mydb["power_measurements"]

    voltage_data = []
    current_data = []

    for x in mycol.find():
        if x['sensor_id'] == 'batterymonitor':
            voltage_data.append(x)
        
    for x in mycol.find():
        if x['sensor_id'] == 'currentmonitor':
            current_data.append(x)
        

    # Get voltage measurements as dataframe
    voltage_data_df = pd.DataFrame(voltage_data)
    voltage_data_df.to_csv('BatteryMonitor.csv', index = False)
    voltage_measurements = voltage_data_df['measurements']

    # Get current measurements as dataframe
    current_data_df = pd.DataFrame(current_data)
    current_data_df.to_csv('CurrentMonitor.csv', index = False)
    current_measurements = current_data_df['measurements']
    
    return voltage_measurements, current_measurements
```

## Voltage DataFrame with timestamps

```python
# Get voltage values

def gen_voltage_df(voltage_measurements, voltage = True):
    
    vol_cap_string = 'voltage'
    if voltage == False:
        vol_cap_string = 'capacity'
    voltage_values = [[]]
    voltage_timestamp = [[]]

    for n in range(len(voltage_measurements)):
        voltage_values.append([])
        voltage_timestamp.append([])

    for n_split in range(len(voltage_measurements)):
        for one_value in voltage_measurements[n_split]:
            voltage_values[n_split].append(one_value['battery2'][vol_cap_string])
            voltage_timestamp[n_split].append(one_value['timestamp'].strftime('%d/%m, %H:%M'))

    # Get mean voltage values
    mean_voltage_values = []
    for split in voltage_values:
        mean_split = np.mean(np.array(split))
        mean_voltage_values.append(mean_split)

    # Get mean timestamp for each split
    mean_voltage_timestamp = []
    for split in voltage_timestamp:
        if len(split) > 0:
            mean_split = split[0]
            mean_voltage_timestamp.append(mean_split)
        else:
            mean_split = None
            mean_voltage_timestamp.append(mean_split)

    # Get voltage DataFrame
    mean_voltage_df = pd.DataFrame(mean_voltage_values, columns = ['mean ' + vol_cap_string])
    mean_voltage_timestamp_df = pd.DataFrame(mean_voltage_timestamp, columns = ['timestamp'])
    voltage_df = pd.concat([mean_voltage_timestamp_df, mean_voltage_df], axis = 1)
    
    # delete last value (=None)
    voltage_df.drop(voltage_df.tail(1).index,inplace=True)
    
    return voltage_df
```

```python
voltage_df = gen_voltage_df(voltage_measurements)
voltage_df
```

## Capacity DataFrame with timestamp

```python
capacity_df = gen_voltage_df(voltage_measurements, voltage = False)
capacity_df
```

## Current DataFrame with timestamps

```python
# Get current values

def gen_current_df(voltage_measurements):
    current_values = [[]]
    current_timestamp = [[]]

    for n in range(len(current_measurements)):
        current_values.append([])
        current_timestamp.append([])

    for n_split in range(len(current_measurements)):
        for one_value in current_measurements[n_split]:
            current_values[n_split].append(one_value['generated']['generatedCurrent'])
            current_timestamp[n_split].append(one_value['timestamp'].strftime('%d/%m, %H:%M'))

    # Get mean voltage values
    mean_current_values = []
    for split in current_values:
        mean_split = np.mean(np.array(split))
        mean_current_values.append(mean_split)

    # Get mean timestamp for each split
    mean_current_timestamp = []
    for split in current_timestamp:
        if len(split) > 0:
            mean_split = split[0]
            mean_current_timestamp.append(mean_split)
        else:
            mean_split = None
            mean_current_timestamp.append(mean_split)


    # Get current DataFrame
    mean_current_df = pd.DataFrame(mean_current_values, columns = ['mean current'])
    mean_current_timestamp_df = pd.DataFrame(mean_current_timestamp, columns = ['timestamp'])
    current_df = pd.concat([mean_current_timestamp_df, mean_current_df], axis = 1)
    
    # delete last value (=None)
    current_df.drop(current_df.tail(1).index,inplace=True)
    
    return current_df
```

```python
# Get begin index
def get_time_index(data_df, start):
    array = data_df.timestamp.str.find(start)+1
    start_index = 0
    for index in array:
        if index == 1:
            break
        start_index = start_index + 1
    return start_index

# Plot voltage data
def plot_data(data_df, start_index = 0,step_size_ticks = 100):
    
    if start_index != 0:
        start_index = get_time_index(data_df, start_index)
        
    ax = data_df.iloc[start_index:].plot(figsize = (10,10), marker = 'h', fillstyle = 'none')
    ax.set_xticks(ticks = np.arange(data_df.index[start_index], data_df.index[-1], step_size_ticks), labels = data_df.timestamp[np.arange(voltage_df.index[start_index], data_df.index[-1], step_size_ticks)]);
    ax.set_xlabel('Time')
    ax.set_title('Measured Data')
```
